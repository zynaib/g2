```python

# Import pandas as pd
import pandas as pd
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
car_sales = pd.read_csv("/content/drive/MyDrive/car-sales.csv")
car_sales.head()
```

```python
# تنظيف وتحويل العمود Price لأرقام
car_sales["Price"] = (
    car_sales["Price"]
    .astype(str)
    .str.replace(r"[^\d]", "", regex=True)   # يشيل أي رموز مش أرقام
    .astype(float)
)

# رسم العمود بعد التنظيف
car_sales["Price"].plot(title="Car Prices")
plt.show()

```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))
car_sales.head()
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column (cumsum())


# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)

# View the car_sales DataFrame
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column (cumsum())
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales))
car_sales.head()
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
print(car_sales.columns)
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
import matplotlib.pyplot as plt

# نتأكد إن العمود Sale Date عبارة عن datetime
car_sales["Sale Date"] = pd.to_datetime(car_sales["Sale Date"], errors="coerce")

# نتأكد إن Total Sales أرقام
car_sales["Total Sales"] = pd.to_numeric(car_sales["Total Sales"], errors="coerce")

# نرسم باستخدام نفس الأعمدة
plt.figure(figsize=(10,5))
plt.plot(car_sales["Sale Date"], car_sales["Total Sales"], marker="o")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales")
plt.title("Sales Growth Over Time")
plt.grid(True)
plt.show()



```
